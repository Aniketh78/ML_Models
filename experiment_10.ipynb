```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.head()
```

```python
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(df))  # Scaling the data
scaled_data.head()

# Visualize the correlation of the scaled data using a heatmap
sns.heatmap(scaled_data.corr(), annot=True)
plt.title('Correlation Matrix of Scaled Data')
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
data_pca = pca.transform(scaled_data)

# Create a DataFrame with PCA components
data_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2', 'PC3'])
data_pca.head()
```

```python
sns.heatmap(data_pca.corr(), annot=True)
plt.title('Correlation Matrix of PCA-transformed Data')
plt.show()
```
